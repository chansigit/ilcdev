```python
import scanpy as sc
```

```python
adata = sc.read_h5ad("/home/csj/mywork/ilc/220913-FL_ABM-joint_comparison/full-alp.final.h5ad")
del adata.layers
```

```python
abm1 = adata[adata.obs.batch=='ABMaLP1'].copy()
abm2 = adata[adata.obs.batch=='ABMaLP2'].copy()
fl   = adata[adata.obs.batch=='FL'].copy()
```

```python
del adata
```

```python
sc.pl.umap(abm1,color='leiden')
sc.pl.umap(abm2,color='leiden')
sc.pl.umap(fl,color='leiden')
```

# abm1 partition 

```python
abm1.obs['leiden_metacell']=abm1.obs['leiden']
sc.tl.leiden(abm1, key_added='leiden_metacell',
             resolution=0.6, restrict_to=('leiden',['0'])
            )
sc.tl.leiden(abm1, key_added='leiden_metacell',
             resolution=0.1, restrict_to=('leiden_metacell',['1'])
            )
sc.tl.leiden(abm1, key_added='leiden_metacell',
             resolution=0.3, restrict_to=('leiden_metacell',['2'])
            )
sc.tl.leiden(abm1, key_added='leiden_metacell',
             resolution=0.3, restrict_to=('leiden_metacell',['3'])
            )
sc.tl.leiden(abm1, key_added='leiden_metacell',
             resolution=0.5, restrict_to=('leiden_metacell',['4'])
            )
sc.tl.leiden(abm1, key_added='leiden_metacell',
             resolution=0.5, restrict_to=('leiden_metacell',['5'])
            )
sc.tl.leiden(abm1, key_added='leiden_metacell',
             resolution=0.5, restrict_to=('leiden_metacell',['6'])
            )
sc.tl.leiden(abm1, key_added='leiden_metacell',
             resolution=0.1, restrict_to=('leiden_metacell',['7'])
            )
sc.tl.leiden(abm1, key_added='leiden_metacell',
             resolution=0.1, restrict_to=('leiden_metacell',['8'])
            )
sc.tl.leiden(abm1, key_added='leiden_metacell',
             resolution=0.1, restrict_to=('leiden_metacell',['9'])
            )
```

```python
abm1= abm1[~abm1.obs['leiden_metacell'].isin(["5,6","0,7","1,3","7,2","3,3","1,4","8,2","1,5"])].copy()
```

# abm2 partition

```python
abm2.obs['leiden_metacell']=abm2.obs['leiden']
sc.tl.leiden(abm2, key_added='leiden_metacell',
             resolution=0.6, restrict_to=('leiden',['0'])
            )
sc.tl.leiden(abm2, key_added='leiden_metacell',
             resolution=0.6, restrict_to=('leiden_metacell',['1'])
            )
sc.tl.leiden(abm2, key_added='leiden_metacell',
             resolution=0.3, restrict_to=('leiden_metacell',['2'])
            )
sc.tl.leiden(abm2, key_added='leiden_metacell',
             resolution=0.3, restrict_to=('leiden_metacell',['3'])
            )
sc.tl.leiden(abm2, key_added='leiden_metacell',
             resolution=0.5, restrict_to=('leiden_metacell',['4'])
            )
sc.tl.leiden(abm2, key_added='leiden_metacell',
             resolution=0.5, restrict_to=('leiden_metacell',['5'])
            )
sc.tl.leiden(abm2, key_added='leiden_metacell',
             resolution=0.05, restrict_to=('leiden_metacell',['6'])
            )
sc.tl.leiden(abm2, key_added='leiden_metacell',
             resolution=0.2, restrict_to=('leiden_metacell',['7'])
            )
sc.tl.leiden(abm2, key_added='leiden_metacell',
             resolution=0.05, restrict_to=('leiden_metacell',['8'])
            )
sc.tl.leiden(abm2, key_added='leiden_metacell',
             resolution=0.01, restrict_to=('leiden_metacell',['9'])
            )
abm2= abm2[~abm2.obs['leiden_metacell'].isin(["5,6","9,1","5,7","8,1","0,8","6,5","6,4","6,3","6,2","0,7","2,3","9,2"])].copy()
```

```python
sc.pl.umap(abm2,color='leiden_metacell', legend_loc='on data')
```

# fl partition

```python
fl.obs['leiden_metacell']=fl.obs['leiden']
sc.tl.leiden(fl, key_added='leiden_metacell',
             resolution=0.6, restrict_to=('leiden',['0'])
            )
sc.tl.leiden(fl, key_added='leiden_metacell',
             resolution=0.1, restrict_to=('leiden_metacell',['1'])
            )
sc.tl.leiden(fl, key_added='leiden_metacell',
             resolution=0.1, restrict_to=('leiden_metacell',['2'])
            )
sc.tl.leiden(fl, key_added='leiden_metacell',
             resolution=0.1, restrict_to=('leiden_metacell',['3'])
            )
sc.tl.leiden(fl, key_added='leiden_metacell',
             resolution=0.2, restrict_to=('leiden_metacell',['4'])
            )
sc.tl.leiden(fl, key_added='leiden_metacell',
             resolution=0.3, restrict_to=('leiden_metacell',['5'])
            )
sc.tl.leiden(fl, key_added='leiden_metacell',
             resolution=0.3, restrict_to=('leiden_metacell',['6'])
            )
sc.tl.leiden(fl, key_added='leiden_metacell',
             resolution=0.2, restrict_to=('leiden_metacell',['7'])
            )
sc.tl.leiden(fl, key_added='leiden_metacell',
             resolution=0.05, restrict_to=('leiden_metacell',['8'])
            )
sc.tl.leiden(fl, key_added='leiden_metacell',
             resolution=0.1, restrict_to=('leiden_metacell',['9'])
            )
sel = fl.obs['leiden_metacell'].value_counts()<=5
fl =  fl[~fl.obs['leiden_metacell'].isin(sel[sel].index.tolist())].copy()
```

```python
sc.pl.umap(fl,color='leiden_metacell', legend_loc='on data')
```

```python
idata= sc.concat([abm1, abm2, fl])
```

```python
idata.obs['batch_leiden_metacell'] = idata.obs['batch'].astype(str)+"_"+idata.obs['leiden_metacell'].astype(str)
```

```python
sc.pl.umap(idata,color='batch_leiden_metacell', legend_loc='on data')
```
