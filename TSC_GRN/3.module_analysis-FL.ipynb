```r
suppressPackageStartupMessages({
    library(ComplexHeatmap)
    library(circlize)
    library(dplyr)
    library(stringr)
    library(pheatmap)
})
o<- function(w,h) options(repr.plot.width=w, repr.plot.height=h)
load(url("https://gitee.com/chansigit/SSAT/raw/master/mm.tf.rda")) #load mm.tf
mm.tf <- c(mm.tf, "Bcl11b")
```

```r
save_pheatmap_pdf <- function(x, filename, width=7, height=7) {
   stopifnot(!missing(x))
   stopifnot(!missing(filename))
   pdf(filename, width=width, height=height)
   grid::grid.newpage()
   grid::grid.draw(x$gtable)
   dev.off()
}
```

```r
'Eomes' %in% mm.tf
```

```r
path='./FL-time_series.csv'
expr_z = read.csv(path,row.names = 1) %>% t %>% scale %>% t
colnames(expr_z) <- 1:ncol(expr_z)

module_info = read.csv('./module_info-FL.csv', row.names = 1)
```

# Gata3 module

```r
# set gene seed
geneSeed="Gata3"

# get module members
(module_info %>% filter(str_detect(member,geneSeed))%>% select(member))[1,]%>% 
str_replace_all(c('\''='', "\\["="", "\\]"="","\n"="")) %>% 
strsplit(' ') %>% unlist -> goi

# build module expression matrix
df = expr_z[goi, ]
isTFann=  data.frame(isTF=factor(ifelse(rownames(df) %in% mm.tf, "yes","no"))) 
rownames(isTFann) <- rownames(df)
isTFcolor= list(isTF = c("yes" = "red", "no" = "white"))

# draw heatmap
o(5,10)
p = pheatmap(expr_z[goi, ], cluster_cols = F, border_color=NA, show_colnames =F,
        breaks=seq(-2, 2, length.out = 101), main=paste(geneSeed, '-Tbx21 module in FL', sep=''),
        annotation_row=isTFann, annotation_colors=isTFcolor)

save_pheatmap_pdf(p, "output/Gata3-Tbx21_module-FL.pdf", width=5, height=10)
```
